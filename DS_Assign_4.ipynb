### Importing Libraries

```python
import numpy as np # linear algebra
import pandas as pd # data processing
import matplotlib.pyplot as plt # plotting library
import missingno as msno # plotting missing data
import seaborn as sns # plotting library
from sklearn.model_selection import train_test_split,cross_val_score, cross_val_predict
from sklearn import metrics
from sklearn.impute import SimpleImputer
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
```

### Import Data

```python
df=pd.read_excel('/content/DS - Assignment Part 1 data set.xlsx')
df.isnull().sum()
df
```

### spliting the Data into X and Y 

```python
x=df.iloc[:,: -1]       
y=df.iloc[:,-1]
```

```python
x_train, x_test, y_train, y_test=train_test_split(x,y,test_size=0.2, random_state=0)
y_Train=list(y_train)
```

### Accuracy of Decision Tree

```python
dt_regressor = DecisionTreeRegressor(random_state = 0)
dt_regressor.fit(x_train,y_train)

#Predicting the Prices using test set 
y_pred_dt = dt_regressor.predict(x_test)

#Decision Tree Regression Accuracy with test set
print('Decision Tree Regression Accuracy: ', dt_regressor.score(x_test,y_test))

#Predicting the Price using cross validation (KFold method)
y_pred_dt = cross_val_predict(dt_regressor, x, y, cv=10 )
#Decision Tree Regression Accuracy with cross validation
accuracy_dt = metrics.r2_score(y, y_pred_dt)
print('Cross-Predicted(KFold) Decision Tree Regression Accuracy: ', accuracy_dt)
```

### Comparing Linear Regression, Decision Tree Regression and RandomForesr Regression

```python
models = []                                                                     
models.append(('LinearRegression',LinearRegression()))
models.append(('DecisionTreeRegressor',DecisionTreeRegressor()))
models.append(('RandomForestRegressor',RandomForestRegressor()))
for name,model in models:
  model.fit(x_train,y_train)
  #Predicting the Prices using test set 
  y_pred=model.predict(x_test)
  #Model Regression Accuracy with test set
  print(name+" : ",model.score(x_test,y_test))
 


```

Importing Amazon and Flipkart Data

```python
amzn= pd.read_csv('/content/amz_com-ecommerce_sample.csv', header= 0, 
                        encoding= 'unicode_escape')   ## amazon Data
flkt=pd.read_csv('/content/flipkart_com-ecommerce_sample.csv',header= 0,  
                        encoding= 'unicode_escape')   ### Flipkart Data
```

```python
amzn.head()
```

```python
retail1=[]                    ## To store retail Price at Flipkart 
name1=[]                      ## to store name of the product at Flipkart
discount1=[]                  ## to store the Discounted Price at Flipkart
retail2=[]                    ## to store retail Price at Amazon
discount2=[]                  ## to store discounted Price at Amazon  
name2=[]                      ## tp store the Name of the Product at Amazon
```

```python
for i in amzn['uniq_id'].tolist():
  fl=flkt[flkt['uniq_id']==i]        ## comparing the product on the basis of uniq_id
  az=amzn[amzn['uniq_id']==i]        ## comparing the product on the basis of uniq_id 
  retail1.append(fl.squeeze()['retail_price'])
  retail2.append(az.squeeze()['retail_price'])
  name1.append(fl.squeeze()['product_name'])
  name2.append(az.squeeze()['product_name'])
  discount1.append(fl.squeeze()['discounted_price'])
  discount2.append(az.squeeze()['discounted_price'])
    
```

```python
# Storing the all the details into Result dataframe
result=pd.DataFrame(list(zip(name1,retail1,discount1,name2,retail2,discount2)),columns=['Productname in Flipkart','Retail Price in Flipkart','Discounted Price in Flipkart','Productname in Amazon','Retail Price in Amazon','Discounted Price in Amazon'])
```
